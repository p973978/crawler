```python
import requests
headers ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.122 Safari/537.36'}
url = 'https://www.dcard.tw/service/api/v2/forums/photography/posts?limit=30&before=233539423'
res = requests.get(url, headers=headers)
print(res.text)

import json

json_str= res.text
json_data = json.loads(json_str)
```

```python
for k in json_data[0]:
    print(k)
```

```python
from urllib import request

#å–å¾—æ¨™æï¼Œç¶²å€ï¼Œåœ–ç‰‡ç¶²å€
for article in json_data:
    article_title = article['title']
    article_url = 'https://www.dcard.tw/f/relationship/p/' + str(article['id'])
    article_img = article['media'] #list
    article_img_list = [i['url'] for i in article['media']]
    
    print(article_title)
    print(article_url)
    print(article_img_list)
    
    for n,img in enumerate(article_img_list):
        img_name = article_title + str(n)+'.' + img.split('.')[-1]
        print(img_name)
        # ./ç•¶å‰ç›®éŒ„
        request.urlretrieve(img, './dcard/{}'.format(img_name))
                                                 
    print('===========')
```
