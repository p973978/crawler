```python
import pandas as pd
import os
```

```python
flist = os.listdir('./ptt_gossiping')
flist
```

```python
columns = ['推','噓','分數','作者','標題','時間'] 
data = []
```

```python
#製作 data frame
# ./ptt_gossiping/ 表示目錄裡面
for f in flist:
    with open('./ptt_gossiping/'+ f, 'r', encoding = 'utf-8') as file:
        tmp_article_content = file.read()
    
    # if the redundent file occur like fw and re.txt, use try to skip it
    # if Directory has error use IsAdirectoryError
    try:
        tmp_data_str = tmp_article_content.split('---split---')[1]
    except IndexError:
        continue
    except IsAdirectoryError:
        continue
    tmp_data = tmp_data_str.split('\n')[1:-1]
    # print(tmp_data)
    
    data.append(tmp_data)
    print(data)
```

```python
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
def column_filter(s):
    output = s.split(': ')[1]
    return output
```

```python
df2 = df.copy() # move to new memory space
```

```python
df
```

```python
df2['推'] = df2['推'].apply(column_filter)
df2['噓'] = df2['噓'].apply(column_filter)
df2['分數'] = df2['分數'].apply(column_filter)
df2['作者'] = df2['作者'].apply(column_filter)
df2['標題'] = df2['標題'].apply(column_filter)
df2['時間'] = df2['時間'].apply(column_filter)
df2
```

```python
for c in columns:
    df[c] = df[c].apply(column_filter)
    
df
```

```python
for c in columns:
    if c != '推':
        df[c] = df[c].apply(lambda s: s.split(': ')[1])
    
df
```
