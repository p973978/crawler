```python
import requests
from bs4 import BeautifulSoup
import os
```

### Get new url by select

```python
headers ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.122 Safari/537.36'}
url = 'https://www.ptt.cc/bbs/Baseball/index.html'

if not os.path.exists('ptt_movie'):
    os.mkdir('ptt_movie')
    
# loop 0, url = 'https://www.ptt.cc/bbs/Baseball/index.html'

for f in range(0,5):
    res = requests.get(url=url, headers=headers)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    title_soup_list = soup.select('div.title')
    for title_soup in title_soup_list:
        title_info = title_soup.select('a')[0]
        title_name = title_info.text
        
        # Extract article content
        article_url = 'https://www.ptt.cc'+title_info['href']
        res_article = requests.get(article_url, headers=headers)
        soup_article = BeautifulSoup(res_article.text, 'html.parser')
        article_content = soup_article.select('div[id="main-content"]')[0].text.split('※ 發信站')[0]
        try:
            with open('./ptt_movie/{}.txt'.format(title_name),'w',encoding='utf-8') as f:
                f.write(article_content)
        except FileNotFoundError:
            title_name = title_name.replace('/','')
            with open('./ptt_movie/{}.txt'.format(title_name),'w', encoding='utf-8') as f:
                f.write(article_content)
        except OSError:
            title_name = title_name.replace('?','')
            with open('./ptt_movie/{}.txt'.format(title_name),'w', encoding='utf-8') as f:
                f.write(article_content)
        
        print(title_name)
        print(article_url)
        print(article_content)
        print('=================')
    
    url = 'https://www.ptt.cc'+ soup.select('a.wide.btn')[1]['href']
    print(url)
```
