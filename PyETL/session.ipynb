```python
import requests
from bs4 import BeautifulSoup

headers ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.122 Safari/537.36'}

url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
ss = requests.session()

res = ss.get(url, headers=headers)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
soup.select('form')
```

```python
over18_url = 'https://www.ptt.cc'+ soup.select('form')[0]['action']
over18_url
```

```python
soup.select('form')[0].select('input')
```

```python
hid_key = soup.select('form')[0].select('input')[0]['name']
hid_key
```

```python
hid_value = soup.select('form')[0].select('input')[0]['value']
hid_value
```

```python
data = {}
data[hid_key] = hid_value
```

```python
data['yes'] = 'yes'
```

```python
ss.cookies
```

```python
ss.post(over18_url, headers=headers, data=data)
```
